```python
#Import liraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
#Import data file
df = pd.read_csv('2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv')

#Adapt date format for daily analysis
df['created_at'] = pd.to_datetime(df['created_at']).dt.date
```

```python
df['order_amount'].describe()
```

```python

%matplotlib inline
plt.style.use('seaborn')

# line plot - daily total amount
df.plot(x='created_at', y='order_amount', color='green', linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend('')

# title and labels
plt.title('Order Amount per date', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Order Amount', fontsize=16)
```

```python
#with a maximum value of 704000$ for one order, we can assume that the Order Amount column contains extreme values.
#These values could be the result of processing errors by shop owners. For instance, shop_id 42 has registered
#several orders where 2000 items were ordered for an amount of 704 000$. Therefore, we can assume that a aive approach 
# to mean calculation will lead to extreme amounts. 
df.loc[(df['shop_id'] == 42)]
```

```python
#Question 1 b):

# Assuming that we do not have enough time to contact the merchant and assess the extreme values issue, 
# we could use the median metric as it is much less sensible to extreme values. This will provide a much more reliable
# number. 
```

```python
#Question 1 c):

print('The order amount median for a 30 day period is: {}'.format(df['order_amount'].median()))
```
